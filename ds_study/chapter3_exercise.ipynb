```python
import pandas as pd

sales = pd.Series(
    [781, 650, 705, 406, 580, 450, 550, 640],
    index=['A','B','C','D','E','F','G','H']
)

print(sales)

sales.size

print(sales.iloc[1])

print(sales.loc['F'])

print(sales.iloc[2:5])

print(sales.loc[['A','B','D']])

print(sales.loc[sales>=600])

print(sales.loc[(sales<500) | (sales>700)])

print(sales.loc[sales>sales['B']])

print(sales.loc[sales<600])
```
