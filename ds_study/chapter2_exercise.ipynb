```python
import pandas as pd
df=pd.DataFrame([['85','96'],
                 ['14','12']])
print(df)
print(df.iloc[1,1])
```

```python
import pandas as pd
score=pd.DataFrame([[85, 96, 40, 95],
                   [73, 69, 45, 80],
                   [78, 50, 60, 90]],
                   index=['John', 'Jane', 'Tom'],
                   columns=['KOR', 'ENG', 'MATH', 'SCI'])
print(score.index)
print(score.columns)
score.iloc[0,2]
print(score.loc['Jane','ENG'])

print(score.iloc[1,'ENG']) -> iloc 함수는 숫자 인덱스만 가능
print(score.loc['Tom',2]) -> loc 은 레이블만 가능
```
